```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/data_pantai/dataset_Google-Maps-Reviews-Scraper_Pantai.csv")
df.head()
```

```python
df = df[['title', 'stars', 'text']]
```

```python
df.info()
```

```python
df['text'] = df['text'].astype(str).str.lower()
df['text'] = df['text'].replace('false', pd.NA)
df = df.dropna(subset=['text'])
df = df[df['text'] != 'false']
```

```python
df = df.drop_duplicates(subset=['text'])
```

```python
df = df.dropna()
```

```python
def clean_gmaps_text(text):
    text = re.sub(r"@[A-Za-z0-9]+", '', text)
    text = re.sub(r"#\w+", '', text)
    text = re.sub(r"RT[\s]+", '', text)
    text = re.sub(r"https?://\S+", '', text)
    text = re.sub(r"[^A-Za-z0-9 ]", '', text)
    text = re.sub(r"\s+", ' ', text).strip()

    return text

df['text'] = df['text'].apply(clean_gmaps_text)
```
